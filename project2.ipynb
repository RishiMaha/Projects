# Section 3: Exploratory Data Visualization


```python
# Visualization to see how many matches took place on each surface to see the most popular type of surface
import matplotlib.pyplot as plt

df["Surface"].hist(bins = 5, edgecolor = 'black')
plt.title('Number of Matches Per Surface')
plt.xlabel('Surface Type')
plt.ylabel('Match Count')
plt.show()
```

This plot is a visual representation showing the distribution of matches by type of surface. The plot is able to illustrate how many matches from the data set were played on each type of surface. It is able to give an understanding of which surfaces are most commonly played on in the mens tour. 

```python
# Filter the 'round column' to show only the finals
finals = df[df['Round'] == 'The Final']

# Count final appearances for each player
finalists = pd.concat([finals['Player_1'], finals['Player_2']])
final_appearances = finalists.value_counts()

# Get the top 10 players with the most final appearances
top_finalists = final_appearances.head(10)


plt.figure(figsize=(10, 6))
top_finalists.plot(kind='bar', color='steelblue')
plt.title('Top 10 Players with Most Final Appearances Since 2013')
plt.xlabel('Player Names')
plt.ylabel('Number of Final Appearances')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This graph shows the 10 players with the most final round appearances since 2013. It shows which players have been dominating the game and successfully competing in tournaments since 2013. 

# Section 4: Data Science


##### How often does the higher ranked player end up winning the match?
I will first clean the data by adding two new columns. One column will be a numerical representaiton of the player with the higher ranking while the other column will be a numerical representation of the player who won the match. This will make it easier to compare the two values.
After cleaning the data, I will find the correlation between the two columns and then calculate the percentage of when the higher ranked player has won the match. 

```python
# Adding another winner column to represent the winner as a number instead of a string
def numeric_winner(row):
    if row['Player_1'] == row['Winner']:
        return 1
    else:
        return 2
df["Num_Winner"] = df.apply(numeric_winner, axis = 1)

# Adding a column to show which player has the higher ranking
def rank(row):
    if row['Rank_1'] < row['Rank_2']:
        return 1
    else:
        return 2
df["Higher_Rank"] = df.apply(rank, axis = 1)
df
```

```python
# Rank comparison
rank_comparison = df["Higher_Rank"].corr(df["Num_Winner"])
print("\nCorrelation between the higher ranked player and the player who won", rank_comparison)
df_same = df[df.Num_Winner == df.Higher_Rank]
prob = len(df_same) / len(df)
print("\nProbability that the higher ranked player was also the winner", prob)
```
